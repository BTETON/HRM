```python
Final=pd.DataFrame([])
```

```python
Final=pd.DataFrame([])
dir="../Data/google_image_matching/0/"
for image in os.listdir(dir):
    if image.endswith(".jpg") : 
        im=Image.open(os.path.join(dir, image))
        A = pd.DataFrame(list(im.getdata()),columns=["R","G","B"])
        B =A.agg(["mean","median","min","max","std"])
        L=pd.DataFrame(B.unstack(),columns=["value"]).reset_index()
        L=L.set_index(L["level_0"]+"_"+L["level_1"])
        Final[image]=L["value"]
```

```python
Final=Final.transpose().reset_index()
```

```python
Final.head()
```

```python
Final["i"]=Final["index"].str.slice(0,5)
Final["j"]=Final["index"].str.slice(6,10)

Final["i"]=pd.to_numeric(Final["i"])
Final["j"]=pd.to_numeric(Final["j"])

#Final = Final.drop('index', 1)
```
